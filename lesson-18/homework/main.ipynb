```python
import pandas as pd
import numpy as np

df = pd.read_csv('tackoverflow_qa.csv')
df.head()
```

```python
# Make sure CreationDate is datetime
df['creationdate'] = pd.to_datetime(df['creationdate'])

# 1. Questions created before 2014
before_2014 = df[df['creationdate'] < '2014-01-01']
print("Questions before 2014:", before_2014.shape[0])

# 2. Questions with score more than 50
score_above_50 = df[df['score'] > 50]
print("Questions with score > 50:", score_above_50.shape[0])

# 3. Questions with score between 50 and 100 (inclusive)
score_50_100 = df[(df['score'] >= 50) & (df['score'] <= 100)]
print("Questions with score between 50 and 100:", score_50_100.shape[0])

# 4. Questions answered by Scott Boston
answered_by_scott = df[df['ans_name'] == 'Scott Boston']
print("Questions answered by Scott Boston:", answered_by_scott.shape[0])

# 5. Questions answered by following 5 users (example users)
users = ['Scott Boston', 'Unutbu', 'Jon Skeet', 'Martijn Pieters', 'Martijn']
answered_by_users = df[df['ans_name'].isin(users)]
print("Questions answered by specified users:", answered_by_users.shape[0])

# 6. Questions created between March 2014 and Oct 2014, answered by Unutbu, with score < 5
mask_date = (df['creationdate'] >= '2014-03-01') & (df['creationdate'] <= '2014-10-31')
mask_answerer = df['ans_name'] == 'Unutbu'
mask_score = df['score'] < 5
filtered = df[mask_date & mask_answerer & mask_score]
print("Questions created between Mar-Oct 2014, answered by Unutbu, score < 5:", filtered.shape[0])

# 7. Questions with score between 5 and 10 OR view count > 10,000
mask_score_range = (df['score'] >= 5) & (df['score'] <= 10)
mask_viewcount = df['answercount'] > 10000
filtered_or = df[mask_score_range | mask_viewcount]
print("Questions with score 5-10 or view count > 10,000:", filtered_or.shape[0])

# 8. Questions NOT answered by Scott Boston
not_answered_by_scott = df[df['ans_name'] != 'Scott Boston']
print("Questions not answered by Scott Boston:", not_answered_by_scott.shape[0])
```
